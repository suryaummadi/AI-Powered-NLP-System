## Loading the Gemini 2.5 Flash

```python
#using model
model = genai.GenerativeModel('models/gemini-2.5-flash')
```

## Developing Function that returns pros and cons in review

```python
def extract_pros_cons_from_review(review_text: str, model) -> dict:

    """
    Given a product review text, uses Gemini API to extract pros and cons
    and returns a Python dictionary with keys 'pros' and 'cons'.
    """
    
    prompt = f"""
Extract pros and cons from the following review. Return ONLY a valid Python dictionary without any markdown or code formatting.

Review: "{review_text}"
"""
    response = model.generate_content(prompt)
    output_text = response.text.strip()

    try:
        output_dict = ast.literal_eval(output_text)
    except Exception as e:
        raise ValueError(f"Failed to parse model output. Output was:\n{output_text}\nError: {e}")

    return output_dict

# Usage:
review = input("Enter your review")
result = extract_pros_cons_from_review(review, model)
print(result)
```
